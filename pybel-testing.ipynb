```python
def convertFPtoBit(smilestring):
    mol = pybel.readstring('smi',smilestring)
    mol.OBMol.AddHydrogens() #make explicit Hs 
    #get list of on bits
    fps=mol.calcfp().bits
    #make bit string from bit list
    temp=list('0'*1024)

    for item in fps:
        temp[item-1]='1'
    #print(''.join(temp)) #bit string in native bit order
    bitstring=''.join(reversed(temp)) #reverse bit order to match openbabel gui    output
    #convert bitstring to hexadecimel
    resultBitStr = []
    for i in range(0, len(bitstring)-1,32):
        resultBitStr.append('%08x'%(int(bitstring[i:i+32],2)) ) #slice 32 bits, convert bin string to int, format as hex padding with zeros
    return resultBitStr
```

```python
import multiprocessing
import time
from functools import partial
def chunks(l, n):
    '''takes a list and integer n as input and returns 
    generator objects of n lengths from that list'''
    for i in range(0, len(l), n):
        yield l[i:i + n]
```

```python
def smrtSrch(chunk,smartString):
    #loop over the id's in the chunk and do the plotting with each
    highlighter = pybel._operations['highlight']
    global matchList
    smart = pybel.Smarts(smartString)
    for smi in chunk:
        mol = pybel.readstring('smi',smi)
        if smart.findall(mol):
            highlighter.Do(mol.OBMol,smrstr+' green')
            matchList.append(mol)
```

```python
cols = ['CID',
 'CanonicalSMILES',
 'InChI',
 'IUPACName',
 'abonds',
 'atoms',
 'bonds',
 'dbonds',
 'formula',
 'HBA1',
 'HBA2',
 'HBD',
 'logP',
 'MP',
 'MR',
 'MW',
 'nF',
 'sbonds',
 'tbonds',
 'TPSA',
 'RotatableBondCount']
```

```python
dfm = pd.DataFrame(columns=cols)
def calcProps(dfm,dfmolec):
    for idx,row in dfmolec.iterrows():      
            mol = pybel.readstring('smi',row['SMILES'])
            desc = mol.calcdesc()
            nrotab = mol.OBMol.NumRotors()
            mform = mol.formula
            desc = {k: v for k, v in desc.items() if pd.Series(v).notna().any()}
            desc['RotatableBondCount'] = nrotab
            desc['formula'] =mform
            try:
                props = pcp.get_properties(['CanonicalSMILES', 'InChI', 'IUPACName'],row['MOLECULE_ID'],'name')[0]
            except IndexError:
                print(f'{idx}: not found')
                pass
            mdict = {**props,**desc}
            df = pd.DataFrame(mdict,index=[0])
            if df.empty:
                print(f'{idx}: no data')
            else:
                print(f'{idx}: success')
            dfm = dfm.append(df,ignore_index=True) 
        
    return dfm
```
