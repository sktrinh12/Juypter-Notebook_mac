```python
import multiprocessing #import Pool, Manager, Process
from bs4 import BeautifulSoup as soup
from urllib.request import urlopen as ureq
import time
```

```python
import csv
combDrgLst_=[]
with open('combDrgLst.txt','r',newline='') as f:
    readcsv=csv.reader(f,delimiter='\n')
    for dr in readcsv:
        combDrgLst_.append(dr)
combDrgLst_ = [x for sl in combDrgLst_ for x in sl]
combDrgLst_
```

```python
searchlink='http://www.chemspider.com/Search.aspx?q='
def f(d,txt):
    name = multiprocessing.current_process().name
    print('Starting',name)
    #d[1].append(4)
    uClient=ureq(searchlink+txt)
    pagehtml = uClient.read()
    soupPage = soup(pagehtml,'html.parser')
    try:
        d[txt] = soupPage.find('td',{'class':'prop_title'},text='#H bond donors:').next_sibling.next_element.text.strip()
    except AttributeError as e:
        print(e)
        d[txt] = 'empty'
    print('Ending',name)
    return d
```

```python
manager = multiprocessing.Manager()
d = manager.dict()
if __name__ == '__main__':    
    #d[1] = []
    for i in range(0,16,4):
        p = multiprocessing.Process(target=f, args=(d,combDrgLst_[i],))
        p2 = multiprocessing.Process(target=f, args=(d,combDrgLst_[i+1],))
        p3 = multiprocessing.Process(target=f, args=(d,combDrgLst_[i+2],))
        p4 = multiprocessing.Process(target=f, args=(d,combDrgLst_[i+3],))
        p.start()
        p2.start()
        p3.start()
        p4.start()
        p.join()
        p2.join()
        p3.join()
        p4.join()
        
print(d)
```
