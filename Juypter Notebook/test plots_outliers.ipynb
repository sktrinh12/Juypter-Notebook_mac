```python
import random
import numpy as np
lst = random.sample(range(350,400),20)
lst.append(950)
```

```python
def reject_outliers(data,m=2):
    return data[abs(data-np.mean(data)) < m*np.std(data)]
```

```python
reject_outliers(np.array(lst))
```
