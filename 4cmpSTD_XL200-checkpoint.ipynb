```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from itertools import cycle

inputfile='M:/_Spencer/BioMicroLabXL200_template/14180806_FINAL.csv'
df = pd.read_csv(inputfile)
cmpds=['Me-4-Hy','Me-4-Hy','Et-4-Hy','Et-4-Hy','Uracil','Uracil','BzPh','BzPh']
cmpdNames= cmpds*int(df.shape[0]/len(cmpds))
df.insert(loc=5,column ='cmpdNames',value=cmpdNames)

fx=[]
srck = df['SrcRack'].astype(str).str.extract('(^\d{9})(\d{2})',expand=False)
srck.columns = ['prefix','suffix']
  
def fillfx(df):
    if df['suffix'] == '11' or df['suffix'] == '12':
        return 'fraction_1'
    elif df['suffix'] == '21' or df['suffix'] == '22':
        return 'fraction_2'
    elif df['suffix'] == '31' or df['suffix'] == '32':
        return 'fraction_3'
    elif df['suffix'] == '41' or df['suffix'] == '42':
        return 'fraction_4'
    elif df['suffix'] == '51' or df['suffix'] == '52':
        return 'fraction_5'
    elif df['suffix'] == '61' or df['suffix'] == '62':
        return 'fraction_6'
    elif df['suffix'] == '71' or df['suffix'] == '72':
        return 'fraction_7'

srck['fx'] = srck.apply(fillfx,axis=1)
srck['netwt'] = df.TgtWGain
srck['cmpdNames'] = df.cmpdNames
srck['replicates'] = [0,1]*int(df.shape[0]/2)
```

```python
srck
```

```python
fig,axs = plt.subplots(2,4,figsize=(15,6))
fig.subplots_adjust(hspace=0.5,wspace=0.2)
axs=axs.ravel()
width=0.25
x = range(1,5)
ymin = min(srck['netwt'])
ymax = max(srck['netwt'])

xtextpos = [x%4 for x in range(1,5)]
for i in range(7):
    fxname='fraction_{}'.format(i+1)
    y1 = srck['netwt'][(srck['fx']==fxname) & (srck['replicates']==0)].tolist()
    y2 = srck['netwt'][(srck['fx']==fxname) & (srck['replicates']==1)].tolist()
    x1 = x
    x2 = [xs+0.25 for xs in x]
    axs[i].bar(x1,y1,width,alpha=0.6,align='center')
    axs[i].bar(x2,y2,width,alpha=0.6,align='center')
    axs[i].set_title(fxname)
    axs[i].set_xticks([xd+0.1425 for xd in x1])
    axs[i].set_xticklabels(cmpds[0::2]) #only first 4 cmpds by taking every other element
    axs[i].set_ylabel('mg')
    axs[i].set_ylim(ymin-width,ymax+width*2)
    axs[i].get_yaxis().set_visible(False)
    patches1 = axs[i].patches[0:8:2] # evens
    patches2 = axs[i].patches[1:8:2] # odds
    labels1 = y1
    labels2 = y2
    cnt=0
    for pat1,lbl1,pat2,lbl2 in zip(patches1,labels1,patches2,labels2):
        height = pat1.get_height()
        axs[i].text(x1[cnt], height+1,lbl1,ha='center',va='bottom',fontsize=8)
        height2 = pat2.get_height()
        axs[i].text(x2[cnt], height2+3,lbl2,ha='center',va='bottom',fontsize=8)
        cnt+=1 
fig.savefig('C:/users/trinhsk/desktop/barplot.png',bbox_inches='tight',dpi=450)
```
