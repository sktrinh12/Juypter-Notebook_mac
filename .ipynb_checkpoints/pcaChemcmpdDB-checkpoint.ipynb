```python
fig, ax = plt.subplots(1,3,figsize=(14,10))

def plotCmpScalr(dfm_numFeatures):
    dfm_std = StandardScaler().fit_transform(dfm_numFeatures) # Gaussian distribution and differing means and standard deviations to a standard Gaussian distribution with a mean of 0 and a standard deviation of 1.
    dfm_norm = Normalizer().fit_transform(dfm_numFeatures) #used if lots of zeros, normalises across rows; all sum to 1
    dfm_mm = MinMaxScaler().fit_transform(dfm_numFeatures)
    lstDf = [dfm_std,dfm_norm,dfm_mm]
    types = ['Std','Norm','MinMax']
    for i,df in enumerate(lstDf):
        features = df.T
        cov_matrix = np.cov(features)
        eigval,eigvec = np.linalg.eig(cov_matrix)
        proj_pc1 = df @ eigvec.T[0]
        proj_pc2 = df @ eigvec.T[1]
        pcadf = pd.DataFrame({'PC1':proj_pc1,'PC2':proj_pc2})
        ax[i].scatter(pcadf.PC1,pcadf.PC2)
        ax[i].set_title = f'{types[i]}'
        
plotCmpScalr(dfm_numFeatures)
        
```

```python
def convertDFtoNumFeatures(dfm):
    dfm_numFeatures = dfm[['csid','cname','amass','logp','hbd','hba','numrotbonds','lrfive','enthalpy','density','psa','bp','arings','numN','numO','sssr','stereoctr','isnp','veberv']]
    dfm_numFeatures.loc[:,'csid'] =dfm_numFeatures['csid'].apply(lambda x : str(x))
    dfm_numFeatures = dfm_numFeatures.set_index('csid')
    clipStr = lambda x: round(float(x.split('±')[0]),1) if x else np.nan
    dfm_numFeatures.loc[:,'enthalpy'] = dfm_numFeatures['enthalpy'].apply(clipStr)
    dfm_numFeatures.loc[:,'density'] = dfm_numFeatures['density'].apply(clipStr)
    dfm_numFeatures.loc[:,'bp'] = dfm_numFeatures['bp'].apply(clipStr)
    dfm_numFeatures.loc[:,'isnp'] = dfm_numFeatures['isnp'].apply(lambda x : 0 if x == 'non-NP' else 1)
    dfm_numFeatures.loc[:,'enthalpy']=dfm_numFeatures['enthalpy'].fillna(round(dfm_numFeatures['enthalpy'].mean(),2))
    dfm_numFeatures.loc[:,'density']=dfm_numFeatures['density'].fillna(round(dfm_numFeatures['density'].mean(),2))
    dfm_numFeatures.loc[:,'bp']=dfm_numFeatures['bp'].fillna(round(dfm_numFeatures['bp'].mean(),2))
    return dfm_numFeatures
```

```python
def genEllipseData (data,p):
    
    '''A sample of independent normally distributed observations, 
    represented as an n-dimensional random vector.
    The random vector can lie anywhere in n-dimensional space, it has n degrees of freedom
    
    Degrees-of-freedom, here a parameter of the distribution, 
    is interpreted as the dimension of an underlying vector subspace.'''
    #s = -2 * log(1 - p) #chi-squared inverse, what extreme value given a probability according to chi-squared distribution
    
    #another way to get the s value is to run percent point function (which is the inverse chi2 fx)
    chisq_stat = chi2.ppf(p,df=data.shape[1]-1) #n-dimension = 2
    
    #rowvar--each column represents a variable, while the rows contain observations. the eigh is for conjugate or real symmetric hermitian arrays
    val, vec = np.linalg.eigh(np.cov(data,rowvar=False)) 
    
    # eigenvalues = variance, which is stdev squared; so take sqrt to get stdev (sigma) 
    width, height = 2 * np.sqrt(val*chisq_stat) 
    theta = np.degrees(np.arctan2(*vec[:,0][::-1])) #first eigenvector set sorted last value first
    return width,height,theta
```

```python
from sklearn.decomposition import PCA


X = iris.data
y = iris.target
#In general a good idea is to scale the data
scaler = StandardScaler()
scaler.fit(X)
X=scaler.transform(X)    

pca = PCA()
x_new = pca.fit_transform(X)

def myplot(score,coeff,labels=None):
    xs = score[:,0]
    ys = score[:,1]
    n = coeff.shape[0]
    scalex = 1.0/(xs.max() - xs.min())
    scaley = 1.0/(ys.max() - ys.min())
    plt.scatter(xs * scalex,ys * scaley, c = y)
    for i in range(n): #two components
        plt.arrow(0, 0, coeff[i,0], coeff[i,1],color = 'r',alpha = 0.5)
        if labels is None:
            plt.text(coeff[i,0]* 1.15, coeff[i,1] * 1.15, "Var"+str(i+1), color = 'g', ha = 'center', va = 'center')
        else:
            plt.text(coeff[i,0]* 1.15, coeff[i,1] * 1.15, labels[i], color = 'g', ha = 'center', va = 'center')
plt.xlim(-1,1)
plt.ylim(-1,1)
plt.xlabel(f"PC{1}")
plt.ylabel(f"PC{2}")
plt.grid()
myplot(x_new[:,0:2],np.transpose(pca.components_[0:2, :]))
t = np.linspace(0,np.pi*2,100)
plt.plot(np.cos(t), np.sin(t), linewidth=1)
plt.show()
```
