```python
'''example taken from site (http://ctr.wikia.com/wiki/Align_the_depiction_using_a_fixed_substructure)'''

query = indigo.loadSmarts("[#7]1~[#6]~[#6]~[#7]~[#6]~[#6]2~[#6]~[#6]~[#6]~[#6]~[#6]~1~2");   #7 is nitrogen, 6 is carbon, ~ is any bond
#SMARTS benzodiazepine system without bond type or atom aromaticity info
coord=[]
collection = indigo.createArray()
refatoms = []

for structure in indigo.iterateSDFile("C:/users/trinh/documents/Python Jupyter/pybel_dplyChemStruc/benzodiazepine.sdf.gz"):
    structure.foldHydrogens() #removal of explicit hydrogen atoms
    match = indigo.substructureMatcher(structure).match(query)
    if structure.index() == 0: #use first molecule as the reference depiction (index =0); all other depictions will have substtructures aligned to the reference
        for atom in query.iterateAtoms():
            coord.extend(match.mapAtom(atom).xyz()) #xyz coordinates of the atom
    else:
        atoms = [match.mapAtom(atom).index() for atom in query.iterateAtoms()]
        structure.alignAtoms(atoms, coord); #alignAtoms - calculate and apply the transformation (scale + rotation + translation) of a molecule so that some subset of its atoms will become as close as possible to the desired positions
 
    refatoms.append(match.mapAtom(query.getAtom(0)).index())
    collection.arrayAdd(structure)
 
    if structure.index() == 15: #only look at first 15 derivaties/analogues of benzodizepine
        break;
filename = 'indigo_grid_white.png'
indigo.setOption("render-output-format", "png");
indigo.setOption("render-image-size", "1000, 1000");
indigo.setOption("render-grid-title-property", "PUBCHEM_COMPOUND_CID");
indigo.setOption("render-grid-title-font-size", "10");
indigo.setOption("render-grid-title-offset", "2");
indigo.setOption("render-coloring", "false");
indigo.setOption("render-background-color", 1.0, 1.0, 1.0);
renderer.renderGridToFile(collection, refatoms, 4, filename)


pil_im = Image.open(filename, 'r')
fig, ax = plt.subplots(figsize=(15,15))
ax.axis('off')
ax.imshow(pil_im, interpolation='nearest')
fig.tight_layout()
```

```python
inputf = 'C:/users/trinh/documents/Python Jupyter/pybel_dplyChemStruc/SummaryPureCmpd.csv'
outpath = 'C:/users/trinh/documents/Python Jupyter/pybel_dplyChemStruc/'
df = pd.read_csv(inputf)
query = indigo.loadQueryMolecule("C(=O)")
collection = indigo.createArray()
for i,r in df.iterrows():
    if r['Structure SMILES'] == 'nan':
        pass
    else:
        try:
            tgt=indigo.loadMolecule(r['Structure SMILES'])
            match = indigo.substructureMatcher(tgt).match(query)
            if match:
                matching = match.highlightedTarget()
                matching.foldHydrogens() # Indigo does not change the representation of the hydrogens automatically
                matching.layout() # if not called, will be done automatically by the renderer
                matching.setProperty("grid-comment", r['Trivial name'])
                collection.arrayAdd(matching)
        except (IndigoException,AttributeError):
            pass
indigo.setOption("render-output-format", "png");
indigo.setOption("render-image-size",8000, 8000);
indigo.setOption("render-bond-length",20)
indigo.setOption("render-comment-position","bottom")
indigo.setOption("render-comment-alignment","center")
indigo.setOption("render-comment-font-size",13)
indigo.setOption("render-comment-offset",2)
indigo.setOption("render-grid-title-property", "grid-comment");
indigo.setOption("render-grid-title-font-size", "25");
indigo.setOption("render-grid-title-offset", "2");
indigo.setOption("render-relative-thickness",1.35);
indigo.setOption("render-label-mode","hetero");
indigo.setOption("render-background-color", 1.0, 1.0, 1.0);
indigo.setOption("render-stereo-style","none");
indigo.setOption("render-margins",1,1);
filename='summaryGridSOM6'
renderer.renderGridToFile(collection,None,7, "{}{}.png".format(outpath,filename))
```

```python
pil_im = Image.open(outpath+filename+'.png', 'r')
fig, ax = plt.subplots(figsize=(200,200))
ax.axis('off')
ax.imshow(pil_im, interpolation='nearest',aspect='auto')
fig.tight_layout()
```
